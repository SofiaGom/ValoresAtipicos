#Procedimiento de desviación estándar para eliminar Outliers

```python
y = df
limite_superior = y.mean() + 3*y.std()
limite_inferior = y.mean() - 3*y.std()
print("Límite superior permitido" , limite_superior)
print("\nLímite inferior permitido" , limite_inferior)
```

```python
#cambiar valores atípicos a nulos
df2 = df[(y <= limite_superior) & (y >=limite_inferior)]
df2
```

```python
df3 = df2.drop(["FECHA", "FOLIO", "UUID", "RFC","PROVEEDOR", "TIPO GASTO", "GASTO","DESCRIPCION","TIPO", "STATUS", "POLIZA"], axis=1)
df3
```

```python
#corroboramos los valores nulos
df3.isnull().sum()
```

```python
#sustituyo con valores de media
df_clean = df3.copy()
df_clean = df_clean.fillna(round(df3.mean(),1))
df_clean
```

```python
df_clean.isnull().sum()
```

#Procedimiento rango intercuantil

```python
y = df_clean

per25 = y.quantile(0.25) #Q1
per75 = y.quantile(0.75) #Q3

iqr = per75 - per75

lim_sup_iqr = per75+ 1.5*iqr
lim_inf_iqr = per25- 1.5*iqr
print("Límite superior permitido" , lim_sup_iqr)
print("\nLímite inferior permitido" , lim_inf_iqr)
```

```python
#datos limpios de df
df3_iqr = df_clean[(y <=lim_sup_iqr)&(y >=lim_inf_iqr)]
df3_iqr
```

```python
df4_iqr = df3_iqr.drop(["IMPORTE"],axis = 1)
df4_iqr
```

```python
nulos = df4_iqr.isnull().sum()
nulos
```

```python
#remplazamos valores atípicos
df5_iqr = df4_iqr.copy()
df5_iqr = df5_iqr.fillna(round(df4_iqr.mean(),1))
df5_iqr
```

```python
#unimos columnas string eliminadas
columnas = ["FECHA", "FOLIO", "UUID", "RFC","PROVEEDOR", "TIPO GASTO", "GASTO","DESCRIPCION","TIPO", "STATUS", "POLIZA"]
indice= df[columnas]
datos_limpios = pd.concat([indice, df5_iqr],axis=1)
datos_limpios
```

```python
#convertir df a csv
datos_limpios.to_csv("Datos limpios.csv")
```
