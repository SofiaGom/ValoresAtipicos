```python
import pandas as pd
df = pd.read_excel("/content/Gastos y costos 20-23.xlsx", sheet_name="2023", header=3)
df
```

```python
df.info()
```

```python
nulos = df.isnull().sum()
nulos
```

#Remplazar valores nulos

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
atras = df["Folio"] =df["Folio"].fillna(method="ffill")
atras2 = df["MP"] =df["MP"].fillna(method="ffill")
atras3 = df["FP"] =df["FP"].fillna(method="ffill")
atras4 = df["TIPO GASTO"] =df["TIPO GASTO"].fillna(method="bfill")
adelante2 = df["Poliza"] =df["Poliza"].fillna(method="bfill")
df = df.drop("Otros", axis=1)
```

```python
nulos = df.isnull().sum()
nulos
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (15,8))
df.plot(kind="box", vert= False)
plt.suptitle("Valores atípicos del excel")
plt.show()
```

#Procedimiento de desviación estándar para eliminar Outliers

```python
y = df
limite_superior = y.mean() + 3*y.std()
limite_inferior = y.mean() - 3*y.std()
print("Límite superior permitido" , limite_superior)
print("\nLímite inferior permitido" , limite_inferior)
```

```python
#cambiar valores atípicos a nulos
df2 = df[(y <= limite_superior) & (y >=limite_inferior)]
df2
```

```python
df3 = df2.drop(["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO", "Descripción","MP", "Tipo", "Status", "Poliza"], axis=1)
df3
```

```python
#corroboramos los valores nulos
df3.isnull().sum()
```

```python
#sustituyo con valores de media
df_clean = df3.copy()
df_clean = df_clean.fillna(round(df3.mean(),1))
df_clean
```

```python
df_clean.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df_clean)
plt.suptitle("Outliers con procedimiento de desviación estandar")
plt.show()
```

#Procedimiento rango intercuantil

```python
y = df_clean

per25 = y.quantile(0.25) #Q1
per75 = y.quantile(0.75) #Q3

iqr = per75 - per75

lim_sup_iqr = per75+ 1.5*iqr
lim_inf_iqr = per25- 1.5*iqr
print("Límite superior permitido" , lim_sup_iqr)
print("\nLímite inferior permitido" , lim_inf_iqr)
```

```python
#datos limpios de df
df3_iqr = df_clean[(y <=lim_sup_iqr)&(y >=lim_inf_iqr)]
df3_iqr
```

```python
nulos = df3_iqr.isnull().sum()
nulos
```

```python
#remplazamos valores atípicos
df5_iqr = df3_iqr.copy()
df5_iqr = df5_iqr.fillna(round(df3_iqr.mean(),1))
df5_iqr
```

```python
#unimos columnas string eliminadas
columnas = [ "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO", "Descripción","MP", "Tipo", "Status", "Poliza"]
indice= df[columnas]
datos_limpios = pd.concat([indice, df5_iqr],axis=1)
datos_limpios
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df5_iqr)
plt.suptitle("Outliers con procedimiento de rango intercuantil")
plt.show()
```

```python
#convertir df a csv
datos_limpios.to_csv("Datos limpios.csv")
```
